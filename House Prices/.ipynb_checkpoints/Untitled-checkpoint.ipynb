```python
numTrain = train.shape[0]
numTest = train.shape[0]
y_train = train.SalePrice.values
all_data = pd.concat((train,test)).reset_index(drop = True)
all_data.drop(['SalePrice'],axis=1,inplace=True)
print(y_train)
print('all data set shape is:{}'.format(all_data.shape))
```

```python
#处理缺失数据
all_missdata = (all_data.isnull().sum()/len(all_data))*100
all_missdata = all_missdata.drop(all_missdata[all_missdata == 0].index)
all_missdata.sort_values(ascending = False)
miss_data = pd.DataFrame({'Miss data':all_missdata})
miss_data.head(10)
```

```python
miss_data.head(20)
```

```python
#相关性检查
corr = train.corr()
plt.figure(figsize=(18, 10))
plt.title('Correlation of House Prices', fontsize=20)
sns.heatmap(corr, annot=False,cmap='RdYlGn', linewidths=0.2)
plt.show()
```

```python
#确实数字数据用0代替
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'MasVnrArea'):
    all_data[col] = all_data[col].fillna(0)
```

```python
#缺失类型数据用none代替
for col in ('PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType', 'MSSubClass'):
    all_data[col] = all_data[col].fillna('None')
```

```python
#最相关的变量缺失数据用众数代替
for col in ('MSZoning', 'Electrical', 'KitchenQual', 'Exterior1st', 'Exterior2nd', 'SaleType', 'Functional', 'Utilities'):
    all_data[col] = all_data[col].fillna(all_data[col].mode()[0])
```

```python
all_data['LotFrontage'] = all_data.groupby('Neighborhood')['LotFrontage'].apply(lambda x: x.fillna(x.median()))
```

```python
all_missdata = (all_data.isnull().sum() / len(all_data)) * 100
all_missdata = all_missdata.drop(all_missdata[all_missdata == 0].index).sort_values(ascending=False)
miss_data = pd.DataFrame({'Missing Ratio':all_missdata})
miss_data.head(30)
#现在没有缺失数据了
```

```python
for col in ('MSSubClass', 'OverallCond', 'YrSold', 'MoSold'):
    all_data[col] = all_data[col].astype(str)
all_data.info()
```

```python
#把正态分布应用到所有歪斜的特征，把斜度大于0.75的值调整为正态分布
numeric_feats = all_data.dtypes[all_data.dtypes != "object"].index
skewed_feats = all_data[numeric_feats].apply(lambda x: skew(x.dropna()))
skewed_feats = skewed_feats[skewed_feats > 0.75]
skewed_feats = skewed_feats.index
all_data[skewed_feats] = np.log1p(all_data[skewed_feats])
```

```python
all_data['TotalSF'] = all_data['TotalBsmtSF'] + all_data['1stFlrSF'] + all_data['2ndFlrSF']
```

```python
low_var_cat = [col for col in all_data.select_dtypes(exclude=['number']) if sum(all_data[col] == mode(all_data[col]))/len(all_data) > 0.97]
low_var_cat
```

```python
all_data = all_data.drop(['Street', 'Utilities', 'Condition2', 'RoofMatl', 'Heating', 'PoolQC'], axis=1)
```

```python
cols = ('FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 
        'ExterQual', 'ExterCond','HeatingQC', 'KitchenQual', 'BsmtFinType1', 
        'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 
        'LandSlope','LotShape', 'PavedDrive', 'Alley', 'CentralAir',
        'MSSubClass', 'OverallCond','YrSold', 'MoSold')

for item in cols:
    lbl = LabelEncoder() 
    lbl.fit(list(all_data[item].values)) 
    all_data[item] = lbl.transform(list(all_data[item].values))

print('Shape all_data: {}'.format(all_data.shape))
```
