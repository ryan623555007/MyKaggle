```python
#numeric 
#是数据平滑
numiric_data = full.columns[full.dtypes != 'object']
numiric_data
```

```python
numeric_means = all_dummies_data.loc[:, numiric_data].mean()
numeric_std = all_dummies_data.loc[:, numiric_data].std()
all_dummies_data.loc[:, numiric_data] = (all_dummies_data.loc[:, numiric_data] - numeric_means) / numeric_std
```

```python
dummies_train = all_dummies_data.loc[train.index]
dummies_test = all_dummies_data.loc[test.index]
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
```

```python
X_train = dummies_train.values
X_test = dummies_test.values
```

```python
alphas = np.logspace(-3,2,50)
test_scores = []
for alpha in alphas:
    clf= Ridge(alpha)
    test_socre = np.sqrt(-cross_val_score(clf,X_train,y_train,cv=10,scoring='mean_squared_error'))
    test_socers.append(np.mean(test_score))
```
